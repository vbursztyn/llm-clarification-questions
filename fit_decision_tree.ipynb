```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier 

import dtreeviz
```

```python
df_data = pd.read_csv('q_and_a_example.csv', index_col=0)

df_data.head(10)
```

```python
y_col = 'city'
X_cols = list(df_data.columns)
X_cols.remove(y_col)

y = df_data.index.values
y_names = df_data[y_col].values
X = df_data[X_cols].values
```

```python
clf = DecisionTreeClassifier(random_state=42)
model = clf.fit(X, y)
```

```python
viz = dtreeviz.model(clf, X, y, target_name='city', feature_names=X_cols, class_names=y_names)

viz.view()
```
